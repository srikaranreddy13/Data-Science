```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
print("tips")
tips.head()
```

```python
print("iris")
iris.head()
```

```python
sns.histplot(tips['total_bill'])
plt.title('Histogram - Distribution of Total Bill')
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.title('Kernel Density Estimation - Distribution of Total Bill')
plt.show()
```

```python
sns.displot(tips['total_bill'],kde='True')
plt.title('Histogram + kde')
plt.show()
```

```python
sns.barplot(x='sex',y='tip',data=tips)
plt.title('Bar Plot - Average Tip by Sex')
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.title('Count Plot - Number of Visitings by Day')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box Plot - Distribution of Total Bill by Day')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin Plot - Distribution of Total Bill by Day')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.title('Scatter Plot - Total Bill vs Tip')
plt.show()
```

```python
sns.lineplot(x='size',y='tip',data=tips)
plt.title('Line Plot -  Tip vs Party size')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',col='day')
plt.title('Relplot - Total Bill vs Tip and gender day by day',y=1.02)
plt.show()
```

```python
iris_num=iris.select_dtypes(float,int)
corr=iris_num.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Heatmap - Correlation between Iris Features')
plt.show()
```

```python
sns.clustermap(corr,annot=True,cmap='coolwarm')
plt.title('Clustermap - Correlation between Iris Features',y=1.02)
plt.show()
```
