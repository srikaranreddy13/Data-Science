Indexing and Slicing:
You are working as a Data Analyst at a fruit distribution company.

The company stores the number of fruit boxes shipped per day (for 1 week) in a 2D NumPy array where:
Each row represents a type of fruit.
Each column represents a day of the week (Mon–Sun).


Your Tasks
Create the 2D NumPy array

Create an array named shipments using the above data.
Access single elements (Indexing):
Get the number of Apples shipped on Wednesday.
Get the number of Oranges shipped on Sunday.
Access multiple elements (Slicing):
Extract all the weekday shipments (Mon–Fri) for all fruits.
Extract only the weekend shipments (Sat & Sun).
Extract the shipments of Apples from Wednesday to Saturday.
Bonus (Practice thinking):
Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.


```python
apples=np.array([50,60,55,70,65,80,90])
oranges=np.array([40,45,50,60,55,70,75])
shipments=np.array([apples,oranges])
print("Apples shipped on wednesday are:",shipments[0,2])
```

```python
print("Oranges shipped on sunday are:",shipments[1,6])
```

```python
print("Mon-Fri\n",shipments[:,0:5])
```

```python
print("Weekend\n",shipments[:,5:7])
```

```python
print("Apples from wednesday to saturday\n",shipments[0:1,2:6])
```

```python
print("Shipments before replacement:",shipments[0,5])
shipments[0,5]=100
print("Shipments after replacement:",shipments[0,5])
```

```python
x=np.array([1,2])
y=np.array([1,1])
print("\n1. Arithmetic ufuncs\n")
print(np.add(x, y))
print(np.subtract(x, y))
print(np.multiply(x, y))
print(np.divide(x, y))
print(np.power(x, y))
print(np.mod(x, y))
print("\n2. Exponential and Logarithmic ufuncs\n")
print(np.exp(x))
print(np.log(x))
print(np.log10(x))
print(np.log2(x))
print(np.expm1(x))
print("\n3. Trigonometric ufuncs\n")
print(np.sin(x))
print(np.cos(x))
print(np.tan(x))
print(np.arcsin(x))
print(np.arccos(x))
print(np.arctan(x))
print(np.degrees(x))
print(np.radians(x))
print("\n4. Rounding ufuncs\n")
print(np.round(x))
print(np.floor(x))
print(np.ceil(x))
print(np.trunc(x))
print("\n5. Statistical ufuncs\n")
print(np.min(x))
print(np.max(x))
print(np.mean(x))
print(np.std(x))
print(np.var(x))
print(np.sum(x))
print(np.prod(x))
print("\n6. Logical ufuncs\n")
print(np.logical_and(x, y))
print(np.logical_or(x, y))
print(np.logical_not(x))
print(np.logical_xor(x, y))
print("\n7. Comparison ufuncs\n")
print(np.greater(x, y))
print(np.less(x, y))
print(np.equal(x, y))
print(np.not_equal(x, y))
print(np.greater_equal(x, y))
print(np.less_equal(x, y))
print("\n8. Bitwise ufuncs\n")
print(np.bitwise_and(x, y))
print(np.bitwise_or(x, y))
print(np.bitwise_xor(x, y))
print(np.invert(x))
```

```python
farmA=np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC=np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
production_all=np.array([farmA,farmB,farmC],dtype=float)
print("Shape of production_all:",production_all.shape)
```

```python
a1=np.arange(10,51)
print(a1)
a2=np.random.rand(3,3)
print(a2)

#array attiributes
print("array attirubets")
print(a1.ndim)
print(a1.shape)
print(a1.size)
print(a1.dtype)

#basic
print("---basic operation---")
print(a1+5)
print(a2*2)

#indexing and slicing
print("--indexing and slicing--")
a3=np.arange(1,26)
a3=a3.reshape(5,5)
print(a3)
print(a3.ndim)
print("first row",a3[0,:])
print("last row",a3[4,:])
print("first 3 rows 2 col",a3[0:3,0:2])

#aggergation
print("--aggergation--")
a4 = np.random.randint(10, 101, size=10)
print(a4)
print("max:",a4.max())
print("min:",a4.min())
print("mean:",a4.mean())
print("sum:",a4.sum())

#reshaping
print("--reshaping--")
a5=np.arange(1,13).reshape(3,4)
print(a5)

#boardcasting
print("--boardingcasting--")
a6=np.ones((3,3),dtype=int)
print(a6)
d1=[1,2,3]
print("after adding")
print(a6+d1)

#filtering
print("--filtering--")
a7=np.random.randint(1,51,size=20)
print(a7)
print(a7[a7%5==0])

#matrix multipilaction
print("--matrix multipilaction--")
a8=np.array([[1,2],[3,4]])
a9=np.array([[5,6],[7,8]])
print(np.dot(a8,a9))

#advanced indexing
print("--fancy indexing--")
a10=np.arange(1,17).reshape(4,4)
print(a10)
r=[0,1,2,3]
c=[0,1,2,3]
print(a10[:,])
print("diagonal elements:",a10[[r],[c]])
```
