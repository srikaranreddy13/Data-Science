```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
top_orders = df.sort_values(by="Revenue", ascending=False)
print("\nTop revenue orders:\n", top_orders.head())
```

```python
product_info = pd.DataFrame({
    "StockCode": ["85123A", "71053", "22423"],
    "Category": ["Kitchen", "Garden", "Electronics"],
    "Supplier": ["Supplier1", "Supplier2", "Supplier3"]
})
```

```python
df = df.merge(product_info, how="left", left_on="StockCode", right_on="StockCode")
print("\nDataset after merge:\n", df.head())
```

```python
country_stats = df.groupby("Country").agg(
    TotalRevenue=("Revenue", "sum"),
    AvgRevenuePerOrder=("Revenue", "mean"),
    OrderCount=("InvoiceNo", "nunique")
).sort_values(by="TotalRevenue", ascending=False)

print("\nCountry-level stats:\n", country_stats.head())
```

```python
price_bins = [0, 5, 20, np.inf]
price_labels = ["cheap", "medium", "expensive"]
df["PriceRange"] = pd.cut(df["UnitPrice"], bins=price_bins, labels=price_labels)

price_stats = df.groupby("PriceRange").agg(
    NumPurchases=("InvoiceNo", "count"),
    AvgRevenue=("Revenue", "mean")
)
print("\nPrice bucket stats:\n", price_stats)
```

```python
numeric_cols = ["Quantity", "UnitPrice", "Revenue"]
correlation = df[numeric_cols].corr()
print("\nCorrelation between numeric columns:\n", correlation)
```
