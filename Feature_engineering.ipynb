— Cleaning Tasks
Convert TotalCharges to numeric + fix missing numeric/categorical values

```python
# Convert TotalCharges to numeric (spaces will become NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# List numeric columns to fill
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Fill numeric NaNs using median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical missing values with "Unknown"
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
df.isnull().sum()
```

— Feature Engineering: tenure_group

```python
def tenure_group(x):
    if x <= 12:
        return "New"
    elif x <= 36:
        return "Regular"
    elif x <= 60:
        return "Loyal"
    else:
        return "Champion"

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
df.head()
```

```python
df.tail()
```

monthly_charge_segment

```python
def charge_segment(x):
    if x < 30:
        return "Low"
    elif x <= 70:
        return "Medium"
    else:
        return "High"

df['monthly_charge_segment'] = df['MonthlyCharges'].apply(charge_segment)
```

```python
df.head()
```

has_internet_service

```python
df['has_internet_service'] = df['InternetService'].map({
    "DSL": 1,
    "Fiber optic": 1,
    "No": 0
})
```

```python
df.head()
```

is_multi_line_user

```python
df['is_multi_line_user'] = df['MultipleLines'].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
df.head()
```

contract_type_code

```python
df['contract_type_code'] = df['Contract'].map({
    "Month-to-month": 0,
    "One year": 1,
    "Two year": 2
})
```

```python
df.head()
```

Drop unnecessary fields

```python
df = df.drop(['customerID', 'gender'], axis=1)
```
