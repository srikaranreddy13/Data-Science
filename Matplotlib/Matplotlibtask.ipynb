```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot(df['total_bill'],df['time'],color='orange',linestyle='--')
plt.xlabel('total_bill')
plt.ylabel('time')
plt.title('total_bill vs time')
plt.show()
```

```python
plt.plot(df['tip'],df['time'],color='orange',linestyle='-')
plt.xlabel('tip')
plt.ylabel('time')
plt.title('tip vs time')
plt.show()
```

```python
plt.bar(df['sex'],df['total_bill'],color='red')
plt.xlabel('total_bill')
plt.ylabel('sex')
plt.title('total_bill vs sex')
plt.show()
```

```python
plt.bar(df['sex'],df['tip'],color='red')
plt.xlabel('tip')
plt.ylabel('sex')
plt.title('tip vs sex')
plt.show()
```

```python
plt.hist(df['total_bill'],bins=10,color='lightgreen',edgecolor='black')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.title('total_bill distribution')
plt.show()
```

```python
plt.hist(df['tip'],bins=10,color='lightgreen',edgecolor='black')
plt.xlabel('tip')
plt.ylabel('frequency')
plt.title('tip distribution')
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'],color='blue',marker='o')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('total_bill vs tip')
plt.show()
```

```python
plt.boxplot(df['total_bill'])
plt.xlabel('total_bill')
plt.title('total_bill boxplot')
plt.show()
```

```python
plt.pie(df['sex'].value_counts(),labels=df['sex'].unique(),autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,4,1)
plt.plot(df['total_bill'],df['time'],color='orange',linestyle='--')
plt.xlabel('total_bill')
plt.ylabel('time')
plt.title('total_bill vs time')

plt.subplot(1,4,2)
plt.plot(df['tip'],df['time'],color='orange',linestyle='-')
plt.xlabel('tip')
plt.ylabel('time')
plt.title('tip vs time')

plt.subplot(1,4,3)
plt.boxplot(df['total_bill'])
plt.xlabel('total_bill')
plt.title('total_bill boxplot')

plt.subplot(1,4,4)
plt.bar(df['sex'],df['total_bill'],color='red')
plt.xlabel('total_bill')
plt.ylabel('sex')
plt.title('total_bill vs sex')

plt.tight_layout()
plt.show()
```
